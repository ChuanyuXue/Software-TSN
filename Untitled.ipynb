```python
import pandas as pd
import numpy as np

import re
```

```python
with open("./baseline/base_uspace.log", 'r') as f:
    data = f.read().split('\n')
```

```python
result = [] 
for x in data:
    result.append(x.replace('.',' ').split(' ')[-2:])
```

```python
varr = []
sec = int(result[0][0])
nano = int(result[0][1])
for x in result[1:]:
    nano += 1e7
    if nano >= 1e9:
        sec += 1
        nano -= 1e9
    varr.append(abs((int(x[0]) - sec)*1e9 + int(x[1]) - nano))
    
```

```python
with open("./etf_sw/etf_uspace.log", 'r') as f:
    data = f.read().split('\n')
result = [] 
for x in data:
    result.append(x.replace('.',' ').split(' ')[-2:])
```

```python
varr2 = []
sec = int(result[0][0])
nano = int(result[0][1])
for x in result[1:]:
    nano += 1e7
    if nano >= 1e9:
        sec += 1
        nano -= 1e9
    varr2.append(abs((int(x[0]) - sec)*1e9 + int(x[1]) - nano))
```
